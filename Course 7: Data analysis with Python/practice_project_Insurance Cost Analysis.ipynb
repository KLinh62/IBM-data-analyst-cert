Handle missing data:

- For continuous attributes (e.g., age), replace missing values with the mean.
- For categorical attributes (e.g., smoker), replace missing values with the most frequent value.
- Update the data types of the respective columns.
- Verify the update using `df.info()`.


````python
#1. For continuous attributes (e.g., age), replace missing values with the mean.
mean_age = df['age'].astype('float').mean(axis=0) #convert data type of mean_age into float
                                                # axis = 0 specifies that the mean should be calculated along the rows
df['age'].replace(np.nan, mean_age, inplace=True)

#2. For categorical attributes (e.g., smoker), replace missing values with the most frequent value.
is_smoker = df['smoker'].value_counts().idxmax() #calculate the most frequent value in column 'smoker'
df['smoker'].replace(np.nan, is_smoker, inplace=True)

#3. Update the data types of the respective columns.
df[['age','smoker']] = df[['age','smoker']].astype('int') #convert to data type: int

#4. Verify the update using df.info()
df.info()
df.head()
````

# Task 4 : Model Development

Fit a linear regression model that may be used to predict the `charges` value, just by using the `smoker` attribute of the dataset. Print the $ R^2 $ score of this model.


````python
lm=LinearRegression() #create a linear regression object

# Define the independent and dependent variable
X = df[['smoker']]
Y = df[['charges']]
# Fit the model
lm.fit(X,Y)
# Print the R^2 score
print(lm.score(X,Y))
````

Create a training pipeline that uses `StandardScaler()`, `PolynomialFeatures()` and `LinearRegression()` to create a model that can predict the `charges` value using all the other attributes of the dataset. There should be even further improvement in the performance.


````python
# Create a pipeline object by createing a list of tuples that contains: name of estimator, model constructor 
# The Input list is passed as an argument, defining the sequence of steps in the pipeline
Input = [('polonomial',PolynomialFeatures(include_bias=False)),('scale',StandardScaler()),('model',LinearRegression())]
# Pipeline constructor
pipe=Pipeline(Input)
# Train the pipeline object
Z = Z.astype(float) #convert data type of all predictors into float
pipe.fit(Z,Y)
yhat = pipe.predict(Z)
print('Using training pipeline, R-squared:', r2_score(Y,yhat)) # r2_score is a function from the sklearn.metrics module
                                        # r2_score compares the predicted values yhat with the true values Y and calculates the R-squared value
````

# Task 5 : Model Refinement

Split the data into training and testing subsets, assuming that 20% of the data will be reserved for testing.


````python
# Note: Z and Y defined above
x_train,x_test,y_train,y_test = train_test_split(Z, Y, test_size=0.2, random_state=1)
````

Initialize a Ridge regressor that used hyperparameter $ \alpha = 0.1 $. Fit the model using training data data subset. Print the $ R^2 $ score for the testing data.


````python
RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train,y_train)
yhat_Ridge = RidgeModel.predict(x_test)
print('Using Ridge regression, R^2 for the testing data is:', r2_score(y_test,yhat_Ridge))
````

Apply polynomial transformation to the training parameters with degree=2. Use this transformed feature set to fit the same regression model, as above, using the training subset. Print the $ R^2 $ score for the testing subset.


````python
# Create a PolynomialFeature object for the transformation
pr=PolynomialFeatures(degree=2)
# Feature transformation
x_train_pr=pr.fit_transform(x_train) # transforms the training data x_train using the PolynomialFeatures object pr
                                    # The fit_transform method performs two steps: 1. Fit: It learns the polynomial features based on the training data.
                                    # 2. Transform: It transforms the training data by creating the new polynomial features.
x_test_pr=pr.fit_transform(x_test) # This line transforms the test data x_test using the already fitted PolynomialFeatures object pr. 
                                # Since pr is already fitted on the training data, it only needs to transform the new data without refitting.
# Train a Ridge Regression model (Ridge Regression adds a penalty term to reduce model complexity and prevent overfitting).
RidgeModel.fit(x_train_pr, y_train) 
y_hat = RidgeModel.predict(x_test_pr)
print(r2_score(y_test,y_hat))
````
